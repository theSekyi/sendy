```python
def convert_dt(df,cols):
    for i in cols:
        df[i] = pd.to_datetime(df[i])
    return df
```

```python
df = convert_dt(df_train,["Arrival at Destination - Time","Arrival at Pickup - Time","Pickup - Time"])
```

```python
df_train["target"] = (df_train["Arrival at Destination - Time"] - df_train["Arrival at Pickup - Time"]).astype('timedelta64[s]')
```

```python
df_train["target_1"] = (df_train["Arrival at Destination - Time"] - df_train["Pickup - Time"]).astype('timedelta64[s]')
```

```python
df_train['compare'] = np.where(df_train['Time from Pickup to Arrival'] == df['target_1'], 'no change', 'changed')
```
